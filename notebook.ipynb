```python
import daft
import matplotlib.pyplot as plt

# ---------- Global style ----------
plt.rcParams.update({
    "figure.dpi": 160,
    "savefig.dpi": 300,
    "font.size": 12,
    "font.family": "DejaVu Sans",
})

# Professional palette (academic/business)
COLORS = {
    "treatment_fill": "#FFF1C9",   # warm pale amber
    "outcome_fill":   "#E8EEF6",   # soft slate-blue
    "stroke":         "#1F2D3D",   # deep navy-gray
    "edge":           "#1F2D3D",
    "title":          "#111827",
}

# ---------- DAG 1 ----------
pgm = daft.PGM(dpi=160, alternate_style="outer")

# Nodes
pgm.add_node(
    "lemons", "Lemons", 1.0, 1.0,
    aspect=2.40, scale=1.25, fontsize=13,
    plot_params={
        "facecolor": COLORS["treatment_fill"],
        "edgecolor": COLORS["edge"],
        "linewidth": 2.4,
        "alpha": 0.98
    }
)

pgm.add_node(
    "scurvy", "Scurvy", 3.0, 1.0,
    aspect=2.40, scale=1.25, fontsize=13,
    plot_params={
        "facecolor": COLORS["outcome_fill"],
        "edgecolor": COLORS["edge"],
        "linewidth": 2.4,
        "alpha": 0.98
    }
)

# Edge
pgm.add_edge("lemons", "scurvy")

# Render
pgm.render()
pgm.figure.set_size_inches(7.0, 2.4)
pgm.ax.set_title("Simplified Lind’s Model (1747)", fontsize=15, pad=14, color=COLORS["title"])

# Clean look (no axis)
pgm.ax.axis("off")
plt.show()

# pgm.figure.savefig("dag1_lind_1747.png", bbox_inches="tight", facecolor="white")
```

```python
import daft
import matplotlib.pyplot as plt

plt.rcParams.update({
    "figure.dpi": 160,
    "savefig.dpi": 300,
    "font.size": 12,
    "font.family": "DejaVu Sans",
})

COLORS = {
    "cause_fill":     "#FDE2E2",  # soft rose
    "mediator_fill":  "#D9F2FF",  # light academic blue
    "outcome_fill":   "#E8EEF6",  # soft slate-blue
    "edge":           "#1F2D3D",
    "title":          "#111827",
}

# ---------- DAG 2 ----------
pgm = daft.PGM(dpi=160, alternate_style="outer")

# Layout chosen to mirror the diagonal reference style
pgm.add_node(
    "acid", "Acid", 1.0, 3.0,
    aspect=2.10, scale=1.25, fontsize=13,
    plot_params={
        "facecolor": COLORS["cause_fill"],
        "edgecolor": COLORS["edge"],
        "linewidth": 2.4,
        "alpha": 0.98
    }
)

pgm.add_node(
    "bacteria", "Bacteria in Food", 2.35, 2.0,
    aspect=3.30, scale=1.25, fontsize=13,
    plot_params={
        "facecolor": COLORS["mediator_fill"],
        "edgecolor": COLORS["edge"],
        "linewidth": 2.4,
        "alpha": 0.98
    }
)

pgm.add_node(
    "scurvy", "Scurvy", 3.60, 1.0,
    aspect=2.40, scale=1.25, fontsize=13,
    plot_params={
        "facecolor": COLORS["outcome_fill"],
        "edgecolor": COLORS["edge"],
        "linewidth": 2.4,
        "alpha": 0.98
    }
)

pgm.add_edge("acid", "bacteria")
pgm.add_edge("bacteria", "scurvy")

pgm.render()
pgm.figure.set_size_inches(7.6, 4.6)
pgm.ax.set_title("Updated Bacterial Theory (Late 19th Century)", fontsize=15, pad=14, color=COLORS["title"])
pgm.ax.axis("off")
plt.show()

# Optional export (uncomment if needed)
# pgm.figure.savefig("dag2_bacterial_theory.png", bbox_inches="tight", facecolor="white")
```
